```python
from gtts import gTTS
import os
    
text = "hello world this is a test of gtts library in python for text to speech conversion and audio file generation and playback "
language = 'en'
output = gTTS(text=text, lang=language, slow=False)
    
output.save("output.mp3")
os.system("mpg321 output.mp3") #for playing the audio
```

```python
from gtts import gTTS
import os


# Text to convert to speech
text = "hello world this is a test of gtts library in python for text to speech conversion and audio file generation and playback" 

# Convert text to speech
tts = gTTS(text=text, lang='en', slow=False,tld='co.in')

# Save the audio file
tts.save("outputlong.mp3")
print("Audio saved as output.mp3")


```

```python
from gtts import gTTS
from datetime import datetime

# Text to convert to speech
text = "Good morning sir this is a test case for sign language"

# Generate filename with current date & time
timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
filename = f"output_{timestamp}.mp3"

# Convert text to speech
tts = gTTS(text=text, lang='en', slow=False)

# Save the audio file
tts.save(filename)

print(f"Audio saved as {filename}")
```
